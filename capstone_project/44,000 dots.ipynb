```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.collections
import matplotlib.patches as mpatches

```

```python
num = 3000
num_2=6
num_3=4
#sizes = 0.2 * np.random.random(num)
xy = 100 * np.random.random((num, 2))
xy_2 = 100 * np.random.random((num_2, 2))
x=100*np.random.random(num_2)
y=100*np.random.random(num_2)
x_3=100*np.random.random(num_3)
y_3=100*np.random.random(num_3)

# Note that the patches won't be added to the axes, instead a collection will
patches = [plt.Circle(center, 1) for center in xy]
patches_2=[plt.Circle(center_2,2) for center_2 in xy_2]
fig, ax = plt.subplots()

coll = matplotlib.collections.PatchCollection(patches, facecolors='yellow')
#coll_2=matplotlib.collections.PathCollection(patches_2,facecolors='red')
ax.add_collection(coll)
#ax.add_collection(coll_2)
target_1=plt.scatter(x,y,color='black')
target_2=plt.scatter(x_3,y_3,color='red')#,linewidth=3)
plt.xlim(0,100)
plt.ylim(0,100)
#ax.margins(0.01)
plt.title('0.3% Recalled, 40% Correctly Identified',fontsize=20)
#plt.xlabel('3,000 dots, 10 recalls')
yellow_patch = mpatches.Patch(color='yellow', label='All Drug Labels')
black_patch = mpatches.Patch(color='black', label='Drug Recalls not Identified')
red_patch = mpatches.Patch(color='red', label='Drug Recalls Identified')

plt.legend(handles=[yellow_patch,black_patch,red_patch])

plt.savefig('correct_id_visual.png')
plt.show()
```
